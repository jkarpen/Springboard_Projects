### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

### B - Perform your own statistical analysis.

To test the validity of the analysis, I plan to run three statistical tests:

 - A linear regression comparing Excess Readmission Ratio to Number of Discharges and find the actual r-squared and p-value. 
 - A t-test comparing the under 100 group mean against the over 1000 group mean
 - A t-test comparing the hospitals with <300 discharges against those with 300 or more. 

#### Test 1 - Linear Regression

```python
#Drop records where Excess Readmission Rate is missing (approx. 81)
no_missing_df = clean_hospital_read_df.dropna(subset=['Excess Readmission Ratio'])
```

```python
#Run linear regression model
#Using scipy.stats.linregress because it provides p-value and r-squared,
#... which np.polyfit does not.
model = scipy.stats.linregress(x=no_missing_df['Number of Discharges'], 
                              y=no_missing_df['Excess Readmission Ratio'])
```

```python
model.pvalue
```

```python
model.rvalue
```

The p-value is extremely small at 1.22e-25, which indicates that this is a statistically significant model. However, the r-squared value is -0.1, which indicates the variables are negatively correlated but to a very small degree (generally an absolute r-squared value of .3 or above is needed to be significant).

While the model is statistically significant, I do not believe it has practical significance due to the low r-squared value. 

#### Test 2 - t-test comparing under 100 discharge hospitals against over 1000 discharge hospitals

The null hypothesis for this test is that the two groups do in fact have the same mean value for Excess Readmission Ratio. 

```python
#Create separate series to analyze
drg_under_100 = no_missing_df[no_missing_df['Number of Discharges'] < 100]['Excess Readmission Ratio']
drg_over_1000 = no_missing_df[no_missing_df['Number of Discharges'] >= 1000]['Excess Readmission Ratio']
```

```python
print(drg_under_100.mean())
print(drg_over_1000.mean())
print(drg_under_100.mean() - drg_over_1000.mean())
```

```python
#Calculate p-value with z-test
ztest1 = ztest(x1=drg_under_100, x2=drg_over_1000)
ztest1
```

The p-value is extremely small, which indicates we can reject the null hypothesis that the mean Excess Readmission Ratio is the same in each group. It seems they are, in fact, different. However the difference in the means is again only 0.044, so in terms of practical significance this does not seem like a very important difference. 

#### Test 3 - t-test comparing under 300 discharge hospitals against 300+ discharge hospitals

```python
#Create separate series to analyze
drg_under_300 = no_missing_df[no_missing_df['Number of Discharges'] < 300]['Excess Readmission Ratio']
drg_over_300 = no_missing_df[no_missing_df['Number of Discharges'] >= 300]['Excess Readmission Ratio']
print(drg_under_300.count())
print(drg_over_300.count())
```

```python
print(drg_under_300.mean())
print(drg_over_300.mean())
print(drg_under_300.mean() - drg_over_300.mean())
```

When comparing the mean Excess Readmission Ratio of hospitals with fewer than 300 discharges against those with more than 300 discharges, we can see the means are very close together, with a difference of only 0.014. This is closer than the difference in means between the under 100 and over 1000 groups.

```python
#Calculate p-value with z-test
ztest2 = ztest(x1=drg_under_300, x2=drg_over_300)
ztest2
```

Once again we have a statistically significant difference with the p-value being extremely small, 5.1e-17. However given that the difference is also extremely small, there is no practical significance to this difference in means. 

### B-4 - Discuss statistical and practical significance

Statistical significance, which has been shown in the tests above, means that in a statistical hypothesis test we have found an outcome that is unlikely to have occurred by chance alone. This is usually shown by a very low p-value. 

Practical significance means that the outcome being analyzed is of a great enough magnitude to be worth spending time or money to address it. Practical significance is not a cut-and-dried metric, but can change case by case depending on what is being analyzed and the opinions of a given stakeholder. In this situation, practical significance means there should be enough of a difference between the two groups to be worth instituting the policy recommendation that all small hospitals need to "demonstrate upgraded resource allocation". 

Given that the majority of hospitals have fewer than 300 discharges, it would cost a great deal of money to implement this plan. The practical significance of doing this for a .014 difference does not seem worth it. 

### B-5 - Discuss the Scatterplot

Scatterplots are often a great way to compare two continuous numeric variables. They allow you to get a sense of the distribution of both variables and see how they relate to each other. In addition they make it easy to see potential outliers in the data.

However, the scatterplot at the top of this exercise is not ideal for this analysis. The data points are far too bunched up at the low end of Number of Discharges and show a mostly flat trend, which the original analyst would have seen if they had added the regression line to the plot.

Instead I would have gone with a boxplot like the first plot below. The second and third box plots attempt to take the analysis further by showing the data by State and Measure Name, respectively, to look for relationships with those variables and Excess Readmission Ratio.


```python
no_missing_df['DRG_Under_300'] = no_missing_df['Number of Discharges'] < 300
```

```python
fig=plt.figure(figsize=(5, 5), dpi=120)

sns.boxplot(x='DRG_Under_300', y='Excess Readmission Ratio', data=no_missing_df)
plt.ylabel("Excess Readmission Ratio")
plt.xlabel("Discharges Under 300")
```

Showing a box-plot here makes it easier to see that there is virtually no difference in the Excess Readmission Ratio between these two groups. The group with under 300 discharges has slightly more outlier values at the high end of Excess Readmission Ratio, but the mean values both hover around 1. 

```python
fig=plt.figure(figsize=(15, 15), dpi=120)

sns.boxplot(x='Excess Readmission Ratio', y='State', data=no_missing_df, 
            order=no_missing_df.groupby('State')['Excess Readmission Ratio'].mean().sort_values().index)
plt.xlabel("Excess Readmission Ratio")
plt.ylabel("State")
```

State appears to be a much more significant determinant of Excess Readmission Ratio. Certain states, like South Dakota and Montana, have very low rates of readmission. Meanwhile states like West Verginia, New Jersey, and the city of Washington D.C. appear to have high rates of readmission. A better recommendation might focus more on the states that are at the high end of the spectrum, and possibly look at the relationship with number of discharges within states. 

```python
fig=plt.figure(figsize=(20, 5), dpi=120)

sns.boxplot(y='Excess Readmission Ratio', x='Measure Name', data=no_missing_df, 
            order=no_missing_df.groupby('Measure Name')['Excess Readmission Ratio'].mean().sort_values().index)
plt.ylabel("Excess Readmission Ratio")
plt.xlabel("Measure Name")
```

Measure Name does not show as much usefulness as State, as all Measures seem to hover around 1 Excess Readmission Ratio.
